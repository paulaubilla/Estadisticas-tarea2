```python
diccionario = {}
for _ , row in df.iterrows():
    BENEF_ENCRIPTADO = row['BENEF_ENCIPTADO']
    MES_INFORMACION = row['MES_INFORMACION']
    TITULAR_CARGA = row['TITULAR_CARGA']
    SEXO_COR = row['SEXO_COR']
    EDAD_TRAMO = row['EDAD_TRAMO']
    COD_REGION_PERNAT = row['COD_REGION_PERNAT']
    DESC_REGION_PERNAT = row['DESC_REGION_PERNAT']
    COD_COMUNA_PERNAT = row['COD_COMUNA_PERNAT']
    DESC_COMUNA_PERNAT = row['DESC_COMUNA_PERNAT']
    NACIONALIDAD_COR = row['NACIONALIDAD_COR']
    TRAMO = row['TRAMO']
    DESC_TIPO_CARACTERIZACION = row['DESC_TIPO_CARACTERIZACION']
    DESC_TIPO_ASEGURADO = row['DESC_TIPO_ASEGURADO']
    SERVICIO_SALUD = row['SERVICIO_SALUD']
    INCRITO_EAPS = row['INCRITO_EAPS']
    DENSIDAD = row['DENSIDAD']
    TRAMO_RENTA = row['TRAMO_RENTA']
    diccionario[BENEF_ENCRIPTADO] = {'MES_INFORMACION': MES_INFORMACION, 'TITULAR_CARGA': TITULAR_CARGA, 'SEXO_COR': SEXO_COR, 'EDAD_TRAMO': EDAD_TRAMO, 'COD_REGION_PERNAT': COD_REGION_PERNAT,
                                      'DESC_REGION_PERNAT': DESC_REGION_PERNAT, 'COD_COMUNA_PERNAT': COD_COMUNA_PERNAT, 'DESC_COMUNA_PERNAT': DESC_COMUNA_PERNAT, 'NACIONALIDAD_COR': NACIONALIDAD_COR,
                                      'TRAMO': DESC_TIPO_CARACTERIZACION, 'DESC_TIPO_CARACTERIZACION': DESC_TIPO_CARACTERIZACION, 'DESC_TIPO_ASEGURADO': DESC_TIPO_ASEGURADO,
                                      'SERVICIO_SALUD': SERVICIO_SALUD, 'INCRITO_EAPS': INCRITO_EAPS, 'DENSIDAD': DENSIDAD, 'TRAMO_RENTA': TRAMO_RENTA}
print(pprint.pprint(diccionario))
```
